For calculating the probability of the (-cs, +ab,+fb), we are here calculating conditional probability formula:


P(-cs, +ab, +fb) = P(-cs | +ab, +fb) * P(+ab) * P(+fb)

**Give a formula (no actual computation coding required) that will derive P(-cs, +ab) from P(-cs, +ab, +fb). Hint: Factor out (eliminate) the fb variable from P(-cs, +ab, +fb) by using a sum using the same method that was used in R2.**




To derive P(-cs, +ab) from P(-cs,+ab,+fb), we are eliminating the fb and using the law of total probability.


Talking about factor out fb from P(-cs,+ab,+fb):
P(-cs, +ab, +fb) = P(+fb) * P(-cs, +ab, +fb | +fb) + P(-fb) * P(-cs, +ab, +fb | -fb)


Where, P(+fb) is the probability that the fan belt is not broken, P(-fb) is the probability that the fan belt is broken.


After that, the sum rule, we are again writing the equation:


P(-cs, +ab, +fb) = P(+fb) * P(-cs, +ab | +fb) + P(-fb) * P(-cs, +ab | -fb)


Where P(-cs, +ab | +fb) this is conditional probability of the car won’t start and alternator is working and fan belt is not broken, and P(-cs, +ab | -fb) this is the conditional probability of the car won’t start and the alternator working with the fan belt broken.
Here , now we are eliminating fb by summing over both possibilities:


P(-cs, +ab) = P(+fb) * P(-cs, +ab | +fb) + P(-fb) * P(-cs, +ab | -fb)


Here we need all the probabilities which are required to calculate P(-cs, +ab) except for P(+fb) and P(-fb). We are calculating the probabilities using the law of total probability.
From fan belt probability table:
P(+fb) = P(fb_y) = 0.7 (fan belt broken)
P(-fb) = P(fb_n) = 0.3 (fan belt not broken)


Here, we are substituting the probabilities of the equation:


P(-cs, +ab) = 0.7 * P(-cs, +ab | +fb) + 0.3 * P(-cs, +ab | -fb)


Above formula is to derive  P(-cs, +ab) from P(-cs, +ab,+fb).


```python
# Define the probability tables
ab = {'ab_y': 0.1, 'ab_n': 0.9}
cs = {}

# Define the joint probability table
for a, p_a in ab.items():
    for c in ['cs_y', 'cs_n']:
        # Initialize the probability to 1
        p = 1
        # Compute the probability based on the given conditions
        if (a == 'ab_y' and c == 'cs_n') or (a == 'ab_n' and c == 'cs_y'):
            p *= p_a
        if (c == 'cs_n' and a == 'ab_y'):
            p *= 0.9
        if (c == 'cs_y' and a == 'ab_n'):
            p *= 0.1
        if c not in cs:
            cs[c] = p
        else:
            cs[c] += p

# Compute the probability of the event (-cs, +ab)
p = cs['cs_n'] * ab['ab_y']
print("P(-cs, +ab) =", p)
```

P(-cs, +fb) = [P(-cs, +ab_y, +fb_y) * P(ab_y) + P(-cs, +ab_n, +fb_y) * P(ab_n)] + [P(-cs, +ab_y, +fb_n) * P(ab_y) + P(-cs, +ab_n, +fb_n) * P(ab_n)]

```python
# Given values
P_bf_bd = 0.8
P_bf_nc = 0.6

# Probability of battery being dead
P_bd_y = 0.2  # from the Battery age table
P_bd_n = 0.8

# Probability of not charging
P_nc_y = 0.4  # calculated from the No charging table
P_nc_n = 0.6

# Probability of fanbelt broken given battery dead
P_bf_bd_y = P_bf_bd * P_bd_y / ((P_bf_bd * P_bd_y) + (P_bf_nc * P_nc_y))

# Probability of fanbelt broken given not charging
P_bf_nc_y = P_bf_nc * P_nc_y / ((P_bf_bd * P_bd_y) + (P_bf_nc * P_nc_y))

if P_bf_bd_y > P_bf_nc_y:
    print("Battery dead is more important in determining whether the battery will go flat.")
elif P_bf_nc_y > P_bf_bd_y:
    print("Not charging is more important in determining whether the battery will go flat.")
else:
    print("Both factors are equally important in determining whether the battery will go flat.")
```
